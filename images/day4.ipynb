```python
class Vector:
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def __add__(self,other):
    return Vector(self.x + other.x,self.y + other.y)
  def __str__(self):
    return (f"vector({self.x}, {self.y})")
  v1 = Vector(1,2)
  v2 = Vector(3,4)
  v3 = v1 + v2
  print(v3)
```

```python
#data abstraction
from abc import ABC,abstractclassmethod
class AbstractDemo:
  def working(self):
    print("we are working")

  @abstractclassmethod
  def studying(self):
   pass

class Concrete(AbstractDemo):
  def studying(self):
    print("we are studying")

obj=Concrete()
obj.studying()
obj.working()
```

```python
class Employee:
  def work(self):
    print("Employee works")

  @abstractclassmethod
  def Develepor(self):
    pass

class Human(Employee):
  @abstractclassmethod
  def sleep(self):
    pass

class FinalClass(Human):
  def developer(self):
    print("we are developers")

  def sleep(self):
    print("human sleeps")

obj=FinalClass()
obj.developer()
obj.sleep()
obj.work()
```

```python
class Shape(ABC):
  def __init__(self):
    print("finding the shape")

  @abstractclassmethod
  def area(self):
    pass

  @abstractclassmethod
  def perimeter(self):
    pass

class Rectangle(Shape):
  def __init__(self,w,h):
    super().__init__()
    self.w=w
    self.h=h

  def area(self):
    return f"Area={self.w*self.h}"

  def perimeter(self):
    return f"perimeter {2*(self.w*self.h)}"


obj=Rectangle(10,5)
print(obj.area())
print(obj.perimeter())
```
